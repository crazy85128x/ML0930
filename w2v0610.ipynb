```
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/mtkdl/raw/master/pttexsmall.csv"
urlretrieve(url, "ptt.csv")
```

```
import pandas as pd
df = pd.read_csv("ptt.csv", encoding="utf-8")
df
```

```
import jieba

url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
urlretrieve(url, "dict.txt.big")
jieba.set_dictionary("dict.txt.big")
url = "https://github.com/Elwing-Chou/mtkdl/raw/master/ptt_dic.txt"
urlretrieve(url, "dict.extra")
jieba.load_userdict("dict.extra")
```

```
import re
def pttcut(content):
    content = re.sub(r'https?:\/\/.*[\r\n]*', '', content)
    punct = set(u''':!),.:;?]}¢'"、。〉》」』】〕〗〞︰︱︳﹐､﹒﹔﹕﹖﹗﹚﹜﹞！），．：；？｜｝︴︶︸︺︼︾﹀﹂﹄﹏､～￠々‖•·ˇˉ―--′’”([{£¥'"‵〈《「『【〔〖（［｛￡￥〝︵︷︹︻︽︿﹁﹃﹙﹛﹝（｛“‘-—_…~/ －＊➜■─★☆=@<>◉é''')
    cut = filter(lambda x: x not in punct, jieba.cut(content))
    return " ".join(cut)
x_train = df["content"].apply(pttcut)
x_train
```

```
with open("pttjoin.txt", "w", encoding="utf-8") as f:
    for c in x_train:
        f.write(c)
        f.write("\n")
```

```
from gensim.models import word2vec

s = word2vec.LineSentence("pttjoin.txt")
# list(s)
model = word2vec.Word2Vec(s)
```

```
model.wv.similarity("正妹", "肥宅")
```

```
import numpy as np
import jieba.analyse
d2v = []
for post in x_train:
    tags = jieba.analyse.extract_tags(post, 
                      topK=None, 
                      withWeight=True)
    dv = np.array([0.0] * 100)
    totalw = 0.0
    for word, tfidf in tags:
        if word in model.wv.vocab:
            dv = dv + tfidf * model.wv[word]
            totalw = totalw + tfidf
    d2v.append(dv / totalw)
```

```
with open("d2v.txt", "w", encoding="utf-8") as f:
    f.write("{} {}\n".format(20000, 100))
    for i, v in enumerate(d2v):
        v = " ".join(map(str, v))
        f.write("{} {}\n".format(i, v))
```

```
from gensim.models import KeyedVectors
d2v_model = KeyedVectors.load_word2vec_format("d2v.txt")
```

```
idx = input("輸入一篇文章代號[0-19999]:")
row = df.iloc[int(idx)]
print("標題:", row["title"])
print("內容:", row["content"])
mostidx, mostcos = d2v_model.wv.most_similar(idx)[0]
print("-" * 50)
row = df.iloc[int(mostidx)]
print("cos:", mostcos)
print("最相似標題:", row["title"])
print("最相似內容:", row["content"])
```
